# Run model

```python
from model import microsim, counterfactual
import transition_matrix as tm
import model_functions as mf
```

```python
with open('steady_noDM_2nd.pickle', 'rb') as handle:
    tp = pickle.load(handle)
tmx = tm.set_transition_matrix(tp)
```

```python
m0 = microsim(tmx, 100000)
states_total_df, cases, ind_state = m0.simulate_population()
```

```python
folder = 'steady_noDM_2nd'
states_total_df.to_csv(folder + '/states_total.csv', index = None)
cases.to_csv(folder + '/cases.csv', index = None)
with open(folder+'/ind_state.pickle', 'wb') as handle:
    pickle.dump(ind_state, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

### see results

```python
dt1, dt2, dt3, dt4, pcmort = mf.get_incidence(np.array(states_total_df), cases['dt1'], cases['dt2'], cases['dt3'], cases['dt4'], cases['pcmort'])
```

```python
c.target_inc.plot()
plt.plot(list(c.age_range), dt1,'b--')
plt.plot(list(c.age_range), dt2, color = 'orange', linestyle= '--')
plt.plot(list(c.age_range), dt3, 'g--')
plt.plot(list(c.age_range), dt4, 'r--')
plt.legend(['t1', 't2', 't3', 't4'], fontsize=15)
plt.xticks(list(range(0,85,5)))
plt.ylabel('Incidence Rate', fontsize = 20)
plt.xlabel('Age', fontsize = 20)
plt.show()
```

```python
plt.plot(c.age_group, c.target_pcmort.values, 'k')
plt.plot(c.age_group, mf.grouping(pcmort), 'k:')
plt.xticks(rotation = 60)
plt.show()
```
