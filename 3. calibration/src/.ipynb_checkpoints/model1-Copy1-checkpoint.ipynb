```python
def set_transition_matrix(params, calibration): # size = number of ages, from state, to state
    '''Creates transition matrix for monthly probabilities
       Input: params = dataframe of all transitions and their probabilities
       Output: numpy matrix
       Size = number of ages x number of states x number of states'''
    
    trans = np.zeros((c.NUM_AGES, c.n_states, c.n_states)) # matrix for each age
    
    step = 0.01
    if calibration == 1:
        
        params_dfs = copy.deepcopy(params)
                 
        # # provide changes once
        # new_params = {}
        # new_trans = cal.change_params(params_dfs[0], step)
        # for age, df in params_dfs.items():
        #     new_params[age] = new_trans
            
        # provide changes to all ages
        new_params = {}
        for age, df in params_dfs.items():
            new_params[age] = cal.change_params(df, step)
        
    else:
        new_params = copy.deepcopy(params)
    # Add initial transition probabilities
    for age in range(c.NUM_AGES):
        for t in c.transition_names:
            start_state = c.p_transition_init.loc[t, 'Start']
            end_state = c.p_transition_init.loc[t, 'End']
            trans[age, c.State[start_state], c.State[end_state]] = use.annual_prob_to_monthly_prob(new_params[age].loc[t, 'Value']) 

    # Add all-cause mortality to all alive states
    for age in range(c.NUM_AGES):
        curr_age = age + 18
        p_acmort = c.acmort.loc[curr_age].item()
        t1_mort = c.tstage_mort['T1'][age]
        t2_mort = c.tstage_mort['T2'][age]
        t3_mort = c.tstage_mort['T3'][age]
        t4_mort = c.tstage_mort['T4'][age]
        
        new_params[age].loc['DT1_PCMORT'] = new_params[age].loc['DT1_PCMORT'].replace(new_params[age].loc['DT1_PCMORT'].Value, t1_mort)
        new_params[age].loc['DT2_PCMORT'] = new_params[age].loc['DT2_PCMORT'].replace(new_params[age].loc['DT2_PCMORT'].Value, t2_mort)
        new_params[age].loc['DT3_PCMORT'] = new_params[age].loc['DT3_PCMORT'].replace(new_params[age].loc['DT3_PCMORT'].Value, t3_mort)
        new_params[age].loc['DT4_PCMORT'] = new_params[age].loc['DT4_PCMORT'].replace(new_params[age].loc['DT4_PCMORT'].Value, t4_mort)

        trans[age, c.State['NORM']:c.State['DT4']+ 1, c.State['ACMORT']] = use.annual_prob_to_monthly_prob(p_acmort)
        trans[age, c.State['DT1'], c.State['PCMORT']] = use.annual_prob_to_monthly_prob(t1_mort)
        trans[age, c.State['DT2'], c.State['PCMORT']] = use.annual_prob_to_monthly_prob(t2_mort)
        trans[age, c.State['DT3'], c.State['PCMORT']] = use.annual_prob_to_monthly_prob(t3_mort)
        trans[age, c.State['DT4'], c.State['PCMORT']] = use.annual_prob_to_monthly_prob(t4_mort)
    
    # Make sure death states have probability of 1 to stay in death state
    trans[:, c.State['ACMORT'], c.State['ACMORT']] = 1
    trans[:, c.State['PCMORT'], c.State['PCMORT']] = 1

    # Calculate stay-in-state probability
    # Add up all out-of-state probabilities for all alive states

    for age in range(c.NUM_AGES):
        for state in c.alive_states:
            if state == 'SURG':
                p_leave = trans[age, c.State[state], :].sum() # sum all probs leaving state
                trans[age, c.State[state], c.State['RECUR']] = trans[age, c.State[state], c.State['RECUR']] + 1 - p_leave
                # print(trans[age, c.State[state], c.State[state]])
            else:
                p_leave = trans[age, c.State[state], :].sum() # sum all probs leaving state
                trans[age, c.State[state], c.State[state]] = 1 - p_leave

    
    # # Make sure probabilities all add to 1
    # # Rows must equal 1
    # for age in range(c.NUM_AGES):
    #     for i_state in range(c.n_states):
    #         if trans[age, i_state].sum() != 1:
    #             print('error')
    #             print(trans[age, i_state].sum())

    return trans, new_params
```

```python
import pickle
with open('steady_withM.pickle', 'rb') as handle:
    p = pickle.load(handle)
```

```python
t_matrix, _ = set_transition_matrix(p, calibration = 0)
```

```python
sojourn_time = []
for t_mat_age in t_matrix:
    s = (1+t_mat_age[1][2]*(1+t_mat_age[2][3]*(1/(1-t_mat_age[3][3])))*(1/(1-t_mat_age[2][2])))*(1/(1-t_mat_age[1][1]))
    sojourn_time.append(s)
```

```python
plt.plot(list(range(18,85)), sojourn_time)
plt.ylabel('sojourn time', fontsize = 12)
plt.xlabel('age', fontsize = 12)
plt.show()
```

```python
def sojourn_measure(tm):
    sj_time = []
    for tm_age in tm:
        s = (1+tm_age[1][2]*(1+tm_age[2][3]*(1/(1-tm_age[3][3])))*(1/(1-tm_age[2][2])))*(1/(1-tm_age[1][1]))
        sj_time.append(s)
    return sj_time

def time2csm(tm):
    m5_10,m6_10,m7_10,m8_10 = [],[],[],[]
    for tm_age in tm:
        m5_10.append(1/(1-tm_age[5][5]))
        m6_10.append(1/(1-tm_age[6][6]))
        m7_10.append(1/(1-tm_age[7][7]))
        m8_10.append(1/(1-tm_age[8][8]))
    return m5_10,m6_10,m7_10,m8_10
```

```python
t1m, t2m, t3m, t4m = time2csm(t_matrix)
```

```python
def gof(obs, exp):
    # chi-squared
    # inputs: numpy arrays of observed and expected values
    chi = ((obs-exp)**2)
    chi_sq = sum(chi)
    return chi_sq.sum()
```
