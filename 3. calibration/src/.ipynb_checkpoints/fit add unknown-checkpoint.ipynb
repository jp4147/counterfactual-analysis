```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
INC_TARGET_PATH = '../data/targets/PC/Pancreatic Cancer SEER Rates.xlsx'
target_incid_by_stages=pd.read_excel(INC_TARGET_PATH, sheet_name='Incidence by stage (single age)')
mid = 40 

loc = target_incid_by_stages.iloc[mid + 3:-3,0:2]
loc.columns=['age','loc']
reg = target_incid_by_stages.iloc[mid + 3:-3, 4:5]
reg.columns=['reg']
dist = target_incid_by_stages.iloc[mid + 3:-3, 7:8]
dist.columns = ['dist']

x_mid=list(range(40,85))
y_loc = list(loc['loc'])
y_reg = list(reg['reg'])
y_dist = list(dist['dist'])
# plt.plot(list(range(18,85)), loc['loc'], 'g')
# plt.plot(list(range(18,85)), reg['reg'], 'r')
# plt.plot(list(range(18,85)), dist['dist'], 'b')
# # plt.plot(list(range(18,85)), target_incid['overall_incid'][:-1], 'k')
x_all= list(range(18,85))

y_loc_all = list(target_incid_by_stages.iloc[21:-3,1:2]['Unnamed: 1'])
y_reg_all = list(target_incid_by_stages.iloc[21:-3,4:5]['Unnamed: 4'])
y_dist_all = list(target_incid_by_stages.iloc[21:-3,7:8]['Unnamed: 7'])
y_uk_all = list(target_incid_by_stages.iloc[21:-3,10:11]['Unnamed: 10'])
```

```python
len(loc)
```

```python
uk=np.array(y_uk_all)/3
y_loc_all = np.array(y_loc_all) + uk
y_reg_all = np.array(y_reg_all) + uk
y_dist_all = np.array(y_dist_all) + uk
```

```python
y_loc = y_loc_all[22:]
y_reg = y_reg_all[22:]
y_dist = y_dist_all[22:]
```

```python
def objective(x, a, b, c):
    return a * x + b * x**2 + c
```

```python
def objective2(x, a, b, c, d):
    return a * x + b * x**2 + c * x**3 + d
```

```python
def func(x, a, b, c):
    return np.exp(x-b) + c
```

```python
popt_loc, pcov = curve_fit(objective, x_mid, y_loc)
a_loc, b_loc, c_loc = popt_loc
# popt_reg, pcov = curve_fit(objective, x, y_reg)
# popt_dist, pcov = curve_fit(objective, x, y_dist)
```

```python
popt_loc, pcov = curve_fit(objective2, x_mid, y_loc)
a_loc, b_loc, c_loc, d_loc = popt_loc

popt_reg, pcov = curve_fit(objective2, x_mid, y_reg)
a_reg, b_reg, c_reg, d_reg = popt_reg

popt_dist, pcov = curve_fit(objective2, x_mid, y_dist)
a_dist, b_dist, c_dist, d_dist = popt_dist
```

```python
print('y = %.10f * x + %.10f * x^2 + %.10f * x^3 + %.10f' % (a_loc, b_loc, c_loc, d_loc))
a_reg, b_reg, c_reg, d_reg = popt_reg
print('y = %.10f * x + %.10f * x^2 + %.10f * x^3 + %.10f' % (a_reg, b_reg, c_reg, d_reg))
a_dist, b_dist, c_dist, d_dist = popt_dist
print('y = %.10f * x + %.10f * x^2 + %.10f * x^3 + %.10f' % (a_dist, b_dist, c_dist, d_dist))
```

# Dist

```python
p=1
px1=40
py1=p
px2=50
py2=-11.4036096587 * px2 + 0.1999768324 * px2**2 + -0.0010604342 * px2**3 + 206.2188293205

s = (py2-py1)/(px2-px1)
b = p-s*px1

y=[]
for age in x_all:
    if age < px1:
        y.append((p-0.1)/(px1-18) * age + p-(p-0.1)/(px1-18)*px1)
    elif age <= px2:
        y.append(s * age + b)
    else:
        y.append(-11.4036096587 * age + 0.1999768324 * age**2 + -0.0010604342 * age**3 + 206.2188293205)

plt.plot(x_all, y_dist_all)
plt.plot(x_all, y, 'r--')
plt.show()
```
