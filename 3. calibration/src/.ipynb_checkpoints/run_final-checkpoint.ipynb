### decreasing mortality

```python
import copy
```

```python
def set_start_state():
    '''Defines starting state for the markov model
       Everyone starts off in NORM'''
    start_state = np.zeros(len(c.state_names))
    start_state[c.states['NORM']] = 1
    return start_state
```

```python
#To apply decreasing transition probability from dt to mort
def withDecreasing_mort(counter, prev_state, prev_dt_history, curr_t_matrix, dF, ut, dt, mort):
    new_dt_history = {}
    dt_pcmort_rev = 0
    
    dt_state = prev_state[c.states[dt]]
    dt_pcmort_trans = curr_t_matrix[c.states[dt], c.states[mort]]
    dt_acmort_trans = curr_t_matrix[c.states[dt], c.states['ACMORT']]
    
    ut_state = prev_state[c.states[ut]]
    ut_dt_trans = curr_t_matrix[c.states[ut], c.states[dt]]
    
    new_pc_cases = 0
    for i in range(counter + 1):
        if i == 0: #new population going into dt and mort
            new_dt_history[i] = ut_state * ut_dt_trans
        else:
            diagnosed_age = np.floor(18 + (counter - 1)/12) # i = survived months
            f = 1
            if i>=12 and i<132: #i==132 is 11 years
                f = dF[diagnosed_age][int(np.floor(i/12))-1]
            else:
                f=0
            new_dt_history[i] = prev_dt_history[i-1] * (1 - dt_pcmort_trans*f - dt_acmort_trans)
            new_pc_cases += prev_dt_history[i-1] * dt_pcmort_trans*f
    
    new_dtcases = new_dt_history[0]
    new_num_dt = sum(new_dt_history.values())
    
    prev_dt_history = copy.deepcopy(new_dt_history)
    
    return new_dtcases, new_num_dt, new_pc_cases, prev_dt_history
```

```python
start_state = set_start_state()

cycle_counter = 0 # updated after transitions are applied
state = {}
new_dt1cases=np.zeros(len(c.age_range))
new_pcmort_cases=np.zeros(len(c.age_range))

for i_age in range(len(c.age_range)):
    # Get current transition matrix, same for every month in the year
    curr_t_matrix = tmx[i_age+18]
    for i_month in range(12):
        if i_age == 0 and i_month == 0: # absolute beginning of the model
            prev_state = start_state
            state[cycle_counter] = prev_state
            prev_num_dt1, prev_num_dt2, prev_num_dt3, prev_num_dt4 = 0, 0, 0, 0
        else:
            prev_state = state[cycle_counter]
            
        new_state = np.dot(prev_state, curr_t_matrix)
        
        ut1_dt1, dt1num, dt1_pc, dt1_history = withDecreasing_mort(cycle_counter, prev_state, prev_dt1_history, curr_t_matrix, d_factor, 'UT1', 'DT1', 'PCMORT')
        ut2_dt2, dt2num, dt2_pc, dt2_history = withDecreasing_mort(cycle_counter, prev_state, prev_dt2_history, curr_t_matrix, d_factor, 'UT2', 'DT2', 'PCMORT')
        ut3_dt3, dt3num, dt3_pc, dt3_history = withDecreasing_mort(cycle_counter, prev_state, prev_dt3_history, curr_t_matrix, d_factor, 'UT3', 'DT3', 'PCMORT')
        ut4_dt4, dt4num, dt4_pc, dt4_history = withDecreasing_mort(cycle_counter, prev_state, prev_dt4_history, curr_t_matrix, d_factor, 'UT4', 'DT4', 'PCMORT')
        
        prev_dt1_history = copy.deepcopy(dt1_history)
        prev_dt2_history = copy.deepcopy(dt2_history)
        prev_dt3_history = copy.deepcopy(dt3_history)
        prev_dt4_history = copy.deepcopy(dt4_history)
        
        new_state_rev = copy.deepcopy(new_state)
        new_state_rev[c.states['DT1']] = dt1num
        new_state_rev[c.states['DT2']] = dt2num
        new_state_rev[c.states['DT3']] = dt3num
        new_state_rev[c.states['DT4']] = dt4num
        new_pc_state = prev_state[c.states['PCMORT']] + dt1_pc + dt2_pc + dt3_pc +dt4_pc
        new_state_rev[c.states['PCMORT']] = new_pc_state
        # new_state_rev[c.states['PCMORT']] = new_pcmort_states
        
        cycle_counter += 1
        state[cycle_counter] = new_state_rev
```
