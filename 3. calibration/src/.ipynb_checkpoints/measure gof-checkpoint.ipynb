```python
import main

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
# Load target data
target_incid_by_stages = pd.read_excel('../data/targets/PC/Pancreatic Cancer SEER Rates.xlsx', sheet_name='Incidence by stage (single age)') # age 18 - 85

# Drop age 85+ and Drop age groups < 1 year, 1-4 years, 5-9 years, 10-14 years
target_loc = target_incid_by_stages.iloc[21:-3,0:2]
target_loc.columns = ['age','rate']
target_reg = target_incid_by_stages.iloc[21:-3, 4:5]
target_reg.columns = ['rate']
target_dist = target_incid_by_stages.iloc[21:-3, 7:8]
target_dist.columns = ['rate']
target_uk = target_incid_by_stages.iloc[21:-3, 10:11]
target_uk.columns = ['rate']

# Convert to numpy array
target_loc = np.array(target_loc['rate'].tolist())
target_reg = np.array(target_reg['rate'].tolist())
target_dist = np.array(target_dist['rate'].tolist())
target_uk = np.array(target_uk['rate'].tolist())

target_loc = target_loc + target_uk/3
target_reg = target_reg + target_uk/3
target_dist = target_dist + target_uk/3
```

```python
def gof(obs, exp):
    # chi-squared
    # inputs: numpy arrays of observed and expected values
    chi = ((obs-exp)**2)
    chi_sq = sum(chi)
    return chi_sq.sum()
```

```python
dat = pd.read_excel('cancer_incidence_loc.xlsx').iloc[:,1:][0:67]

loc = np.array(dat['loc'])
reg = np.array(dat['reg'])
dist = np.array(dat['dist'])
```

```python
gof(loc, target_loc)
```

```python
gof(target_loc, loc)
```

```python
gof(reg, target_reg)
```

```python
gof(dist, target_dist)
```

```python
plt.plot(target_loc, 'g-')
plt.plot(target_reg, 'r-')
plt.plot(target_dist, 'b-')

plt.plot(loc, 'g--')
plt.plot(reg, 'r--')
plt.plot(dist, 'b--')
plt.show()
```

```python
import pickle
with open('gofs_locreg50.pickle', 'rb') as handle:
    iter50 = pickle.load(handle)
    
with open('gofs_locreg200.pickle', 'rb') as handle:
    iter200 = pickle.load(handle)
    
iter50_df = pd.DataFrame(iter50)
iter50_df.columns = ['loc','reg','dist']

iter200_df = pd.DataFrame(iter200)
iter200_df.columns = ['loc','reg','dist']

plt.figure(figsize=(12,4))

for i in range(3):
    plt.subplot(1,3,i+1)
    c = iter50_df.columns[i]
    plt.plot(iter50_df[c])
    plt.plot(iter200_df[c])
    plt.title(c)
    plt.legend(['iter50', 'iter200'])

plt.tight_layout ()
plt.show()
```
