# Run model

```python
from model import markov
import transition_matrix as tm
import model_functions as mf
```

```python
with open('var_t4t3.pickle', 'rb') as handle:
    tp = pickle.load(handle)
    
# new_tp = {}
# for k, v in tp.items():
#     new_tp[k+18] = v
tmx = tm.set_transition_matrix(tp)
```

```python
m0 = markov(tmx)
states_total, cases = m0.run_markov()
```

### see results

```python
dt1, dt2, dt3, dt4, pcmort = mf.get_incidence(np.array(states_total), cases['dt1'], cases['dt2'], cases['dt3'], cases['dt4'], cases['pcmort'])
```

```python
c.target_inc.index = list(c.age_range)
c.target_inc.plot()
plt.plot(list(c.age_range), dt1,'b--')
plt.plot(list(c.age_range), dt2, color = 'orange', linestyle= '--')
plt.plot(list(c.age_range), dt3, 'g--')
plt.plot(list(c.age_range), dt4, 'r--')
plt.legend(['t1', 't2', 't3', 't4'], fontsize=15)
plt.xticks(list(range(15,85,5)))
plt.ylabel('Incidence Rate', fontsize = 20)
plt.xlabel('Age', fontsize = 20)
plt.show()
```

```python
plt.plot(c.age_group, c.target_pcmort.values, 'k')
plt.plot(c.age_group, mf.grouping(pcmort), 'k:')
plt.xticks(rotation = 60)
plt.ylabel('Mortality')
plt.show()
```

```python
plt.plot(np.array(mf.sojourn_time(tmx)))
time36 = np.array(mf.sojourn_time(tmx))
plt.plot(c.target_sojourn,'r:')
noise = np.random.normal(0, 0.5, time36.shape)
new_signal = time36 + noise
plt.plot(new_signal, 'k:')
```

```python
with open('time36.pickle', 'wb') as handle:
    pickle.dump(new_signal, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
plt.plot(mf.cancer_progression(tmx))
time12 = np.array(mf.cancer_progression(tmx))
plt.plot(c.target_cancer_progression,'r:')
noise = np.random.normal(0, 0.5, time12.shape)
new_signal = time12 + noise
plt.plot(new_signal, 'k:')
```

```python
with open('time12.pickle', 'wb') as handle:
    pickle.dump(new_signal, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

# Calibration

```python
import configs as c

from timeit import default_timer as timer

from datetime import timedelta

from datetime import datetime

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
from calibration import sim_anneal
```

```python
start = timer()
cal_markov = sim_anneal(tp_path = 'var_t1.pickle', age_var = 1, model = 'mk', save = 'var_t1_t2.pickle')
cal_tp = cal_markov.anneal()
end = timer()
print(f'total time: {timedelta(seconds=end-start)}')
```

```python
with open('cal_pcmort_time_pcmort_t4.pickle', 'wb') as handle:
    pickle.dump(cal_tp, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
