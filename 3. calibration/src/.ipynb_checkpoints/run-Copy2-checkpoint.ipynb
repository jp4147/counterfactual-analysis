```python
def cancer_progression(tm):
    sj_time = []
    for k, tm_age in tm.items():
        s = (1+tm_age[1][2]*(1+tm_age[2][3]*(1/(1-tm_age[3][3])))*(1/(1-tm_age[2][2])))*(1/(1-tm_age[1][1]))
        sj_time.append(s)
    return sj_time

def sojourn_time(tm):
    sj_time = []
    for k, tm_age in tm.items():
        s = np.mean([1/(1-tm_age[1][1]), 1/(1-tm_age[2][2]), 1/(1-tm_age[3][3]), 1/(1-tm_age[4][4])])
        sj_time.append(s)
    return sj_time
```

```python
ut1_ut2 = 0.9
ut2_ut3 = 0.9
ut3_ut4 = 0.9

ut1_dt1 = 0.6
ut2_dt2 = 0.6
ut3_dt3 = 0.3
ut4_dt4 = 0.1

acmort = 0.0006
ut3_ut3 = 1-ut3_ut4-ut3_dt3-acmort
ut2_ut2 = 1-ut2_ut3-ut2_dt2-acmort
ut1_ut1 = 1-ut1_ut2-ut1_dt1-acmort
ut4_ut4 = 1-ut4_dt4-acmort
```

```python
(1+ut1_ut2*(1+ut2_ut3*(1/(1-ut3_ut3)))*(1/(1-ut2_ut2)))*(1/(1-ut1_ut1)) #1
```

```python
np.mean([1/(1-ut1_ut1), 1/(1-ut2_ut2), 1/(1-ut3_ut3), 1/(1-ut4_ut4)]) #3
```
