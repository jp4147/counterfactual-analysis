```python
from scipy.stats import kde
```

```python
mu, sigma = 60, 10 # mean and standard deviation
s = np.random.normal(mu, sigma, 82)
```

```python
plt.plot(s)
```

```python
density_function = kde.gaussian_kde(s, bw_method=0.3)
# x = np.linspace(0, len(s), len(s))
x = np.array(list(range(18,85)))
 
plt.plot(x, density_function(x))
plt.show()
```

```python
maxv = max(density_function(x))
minv = min(density_function(x))
plt.plot(x, [float(i)/(maxv) for i in density_function(x)])
```
